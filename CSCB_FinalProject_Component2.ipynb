Yes, you can run regress_out before doing clustering. But if you did cell type annotation already, you can actually tell if it worked without re-doing your cell type annotation, because you can check if any clusters have both 10X and PIP-seq data. If each cluster has only 10X or only PIP-seq, then regress_out won't work, and you don't have to pursue it any further. It's possible that it won't work: this is a common method that people use to get rid of unwanted sources of variation, but it may not be flexible enough for this particular case. The reason to try it anyway is that when you use complicated analyses, someone will usually want to know why you didn't start with a simpler tactic.

It replaces each gene's expression with the residuals from a linear regression on the user-provided metadata. It works like a batch correction method, and it's the first and easiest option to reduce systematic differences between two groups. It can serve as a baseline to these other methods if it's inserted before running PCA and clustering. The hope is that the cluster labels will be less driven by whatever factor was regressed out
